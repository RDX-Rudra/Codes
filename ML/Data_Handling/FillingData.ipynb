```python
df.fillna(method="bfill")
```

"bfill" = backword filling, fills data from below row or before column
"ffill" = forword filling, fill data from above row or after column
axis = 1 means columns wise and axis = 0 means row wise

```python
df.fillna(method="ffill", axis=1)
```

Using mode methode

```python
df.isnull().sum()
```

```python
gMode = df["emp_length"].mode()
df["emp_length"].fillna(gMode[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.select_dtypes(include="object")
```

```python
df.select_dtypes(include="object").isnull().sum()
```

```python
for col in df.select_dtypes(include='object').columns:
    mode = df[col].mode()[0]  # Calculate the mode
    df[col] = df[col].fillna(mode)  # Fill missing values with the mode
```

float64 type data handling

```python
from sklearn.impute import SimpleImputer 
```

```python
df.select_dtypes(include="float64").columns
```

```python
si = SimpleImputer(strategy="mean")
arr = si.fit_transform(df[['int_rate', 'emp_length', 'annual_inc', 'dti', 'delinq_2yrs','revol_util', 'total_acc', 'longest_credit_length']])
```

```python
new_df = pd.DataFrame(arr,columns=df.select_dtypes(include="float64").columns)
```

```python
new_df.head()
```

```python
new_df.isnull().sum()
```

```python
new_df['int_rate'].mean()
```

```python
new_df['dti'].mean()
```

```python
for col in df.select_dtypes(include="float64").columns:
    df.fillna({col: new_df[col].mean()}, inplace=True)
```
