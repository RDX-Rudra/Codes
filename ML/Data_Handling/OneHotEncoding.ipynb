One Hot Encoding

```python
import pandas as pd
```

```python
df = pd.read_csv('loan.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
en_df = df[["verification_status"]]
en_df.head()
```

```python
pd.get_dummies(en_df).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()

arr = ohe.fit_transform(en_df).toarray()
```

```python
pd.DataFrame(arr, columns=["verification_status_1", "verification_status_2"])
```

transfer two columns to one columns

```python
ohe = OneHotEncoder(drop="first")
arr = ohe.fit_transform(en_df).toarray()
arr
```

```python
pd.DataFrame(arr, columns=["verification_status"])
```
